```python
import re

for i, row in enumerate(customers, start=0):
    errors = []

    if not re.fullmatch(r"[가-힣]{2,5}", row["이름"]):
        errors.append("이름")

    if not re.fullmatch(r"010-\d{4}-\d{4}", row["전화"]):
        errors.append("전화번호")

    if not re.fullmatch(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}", row["메일"]):
        errors.append("이메일")

    if not re.fullmatch(r"\d{5}", row["우편번호"]):
        errors.append("우편번호")

    if not re.fullmatch(r"\d{4}-\d{2}-\d{2}", row["생년월일"]):
        errors.append("생년월일")

    if not re.fullmatch(r"(남|여)", row["성별"]):
        errors.append("성별")

    if errors:
        print(f"인덱스[{i}] 데이터")
        print("오류:", ", ".join(errors))
        print()
```

1. 문자열에서 숫자를 모두 찾아 출력하시오.
- 출력:  ['123', '456']

```python
import re

text = "Python123AI456"

print(re.findall(r"\d+", text))
```

2. 문자열에서 소문자만 모두 출력하시오.
- 출력: ['y', 't', 'h', 'o', 'n']

```python
text = "PythonAI"

print(re.findall(r"[a-z]", text))
```

3. 문자열에 숫자가 포함되어 있는지 검사하시오.
- 출력: 숫자가 있습니다.

```python
text = "Python123"

if bool(re.search(r"\d+", text))==0:
    print("숫자가 없습니다")
else:
    print("숫자가 있습니다.")
```

4. 문자열에서 전화번호를 찾아 출력하시오.
- 출력: 010-1234-5678

```python
text = "문의 : 010-1234-5678"

print(re.search(r"\d{3}-\d{4}-\d{4}", text).group())
```

5. 문자열에서 이메일 찾아 출력하시오.
- 출력: python@test.com

```python
text = "문의 : python@test.com"

email = re.search(r"\w+@\w+\.\w+", text)

print(email.group())
```

6. 문자열에서 숫자를 모두 *로 변경하시오.
- 결과: A** B** C**

```python
text = "A10 B20 C30"

print(re.sub(r"\d", "*", text))
```

7. 문자열에서 HTML태그("<"로 시작하고 ">"로 끝나는 문자)를 제거하시오.
- 출력: Python

```python
html = "<h1>Python</h1>"

print(re.sub(r"<.*?>", "",html))
```

8. 문자열에서 날짜를 찾아 출력하시오.
- 출력: ['2026-07-07', '2026-08-01']

```python
text = """
오늘은 2026-07-07입니다.
시험은 2026-08-01입니다.
"""

print(re.findall(r"\d+-\d+-\d+", text))
```

9. 아이디는 영문 소문자와 숫자로만 이루어진 5~10글자이다.
다음 아이디중 조건에 맞는 것만 출력하시오.

- 출력:
python
abc123

```python
ids = [
    "python",
    "abc123",
    "ABC123",
    "py",
    "python12345"
]

for t in ids:
    if(bool(re.findall(r"a-z0-9{5:10}, t"))) != 0:
        print(t)
```
