```python
import re

text = "Python"
result = re.match(r"th", text)
print(result) # None match()는 문자열의 시작에서만 검사
print(result.group())
```

```python
text = "apple banana orange"
result = re.findall(r"a", text)
print(result) # ['a', 'a', 'a', 'a', 'a']
result2 = re.findall(r"[an]", text)
print(result2) # ['a', 'a', 'a', 'a', 'a']
```

```python
text = "A10 B20 C30"
result = re.sub(r"\d", "", text)
print(result)
result2 = re.sub(r"\D", "0", text)
print(result2)
result3 = re.sub(r"[^\d\s]", "", text)
print(result3)
```

```python
products = [
    "사과 3500원",
    "배 4200원",
    "수박 18000원"
]

price_list=[]

for product in products:
    name = re.search(r"[가-힣]+", product)  #search 대신 match써도 됨, "^[가-힣]+" 써도 돼
    price = re.search(r"\d+", product)      #얜 match 쓰면 안 돼
    price_list.append({name.group(): int(price.group())})

print(price_list)
# [{"사과":3500}, {"배": 4200}, {"수박": 18000}]
```

```python
emails = [
    "python@test.com",
    "python@네이버.com",
    "kim123@gmail.com",
    "abc@company.co.kr",
    "hello",
    "@gmail.com"
]

pattern = r"^[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}$"
# [A-Za-z0-9._%+-]로 시작
# +:여러자 와도 돼
# @[A-Za-z0-9.-]: 골뱅이로 시작해서 이런 문자가 와야함
# \.[A-Za-z]{2,}$: 그냥 .은 한글자를 대신함. \.: .으로 시작해야한다. 
# {2,}$: .으로 시작해서 2자 이상으로 끝나야한다.

for email in emails:
    if re.fullmatch(pattern, email):
        print(email, ": 정상")
    else:
        print(email, ": 오류")
```

```python
text = "1,23,456,7890,77a8888"
print(re.findall(r"\d{3}", text)) # 숫자문자가 3번연속 나오는 구간만 반환 리스트에 추가
# ['456', '789', '888']
```

```python
text = "1 12 123 1234 12345"

print(re.findall(r"\d{2,4}", text)) # ['12', '123', '1234', '1234']
```

```python
text = "cat, dog, bird, cats, dog"

print(re.findall(r"cat|dog", text)) # ['cat', 'dog', 'cat', 'dog']
print(re.findall(r"[(cat)(dog)]", text))   #(,c,a,t,),(,d,o,g)
```

```python
phones = [
    "02-1234-5678",
    "010-1234-5678",
    "010-123-5678",
    "010-1234-11111"
]

# "숫자3시작-숫자4-숫자4끝나는" 찾기
for p in phones:
    result = re.search(r"^\d{3}-\d{4}-\d{4}$", p)
    if result != None:
        print(result)
        print(result.group())
```

```python
urls = [
    "https://www.google.com",
    "http://naver.com",
    "https://python.org/docs",
    "google.com"
]

pattern = r"^https?://([\w-]+\.)+[\w-]+(/[^\s]*)?$"
# https?: http나 https로 시작해야 함.

for url in urls:
    if re.fullmatch(pattern, url):
        print(url, ": 정상")
    else:
        print(url, ": 오류")
```
