```python
# 웹 브라우저 "뒤로가기" 구현
stack = []

stack.append("네이버")
stack.append("구글")
stack.append("ChatGPT")

print("현재 방문 기록 :", stack)
print("뒤로가기 :", stack.pop())    #뒤로가기 버튼 클릭 이라면
print("현재 페이지 :", stack[-1])   #pop연산 후 top연산
```

```python
# "접시(동전) 쌓기" 구현

stack = []

stack.append("접시1")
stack.append("접시2")
stack.append("접시3")

while len(stack) > 0:
    print(stack.pop())
```

```python
plates = []
plates.append({"color": "red", "price":100})
plates.append({"color": "blue", "price":150})
plates.append({"color": "blue", "price":150})
plates.append({"color": "black", "price":200})
plates.append({"color": "red", "price":100})

print(plates)

total_price = 0

while len(plates)>0:
    p = plates.pop()
    total_price += p["price"]
print(f"{total_price}엔")
print(plates)
```
