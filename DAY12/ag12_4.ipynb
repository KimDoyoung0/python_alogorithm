```python
queue = ["A", "B", "C"]
pop_index = 0
while pop_index < len(queue):
    print(queue[pop_index]) # ["A", "B", "C"]
    print(queue[pop_index+1:])
    pop_index += 1

print(queue)
```

```python
from collections import deque

queue = deque()

queue.append("A")
queue.append("B")
queue.append("C")

print(queue.popleft())
print(queue)
print(queue.pop())
print(queue)
```
