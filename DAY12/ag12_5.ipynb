```python
from collections import deque

dq = deque()
dq.append(10)
dq.append(20)
dq.append(30)

print(dq)
```

```python
dq = deque()
dq.append(30)
dq.append(20)
dq.append(10)

print(dq)
```

```python
dq = deque([10, 20, 30])
x=dq.pop()      #스텍처럼

print(x)
print(dq)
```

```python
dq = deque([10, 20, 30])
x = dq.popleft()    #큐처럼

print(x)
print(dq)
```

스텍이 필요-> 효율성 Deque x -> list 사용
큐 필요 -> 효율성 O -> deque 사용
덱 필요 -> deque 사용
자료구조의 회전 -> 시계/빈시계 방향 (=양방향) 삽입/삭제 가능 -> 

```python
dq = deque()

dq.append(10)
dq.append(20)
dq.appendleft(5)
dq.appendleft(1)

print(dq)

print(dq.pop())
print(dq.popleft())

print(dq)
```

```python
dq = deque([1, 2, 3, 4, 5])
dq.rotate(1) # 우측으로 한칸 이동(회전)
print(dq) # deque([5, 1, 2, 3, 4])

dq.rotate(2) # 우측으로 두칸 이동(회전)
print(dq)

dq.rotate(-1) # 반시계 방향으로 한칸 이동(회전)
print(dq)

###맨 위의 코드와 동일한 코드
dq1 = deque([1,2,3,4,5])
x=dq1.pop()
dq1.appendleft(x)
print(dq1)
```

기본 모듈 안에 덱을 사용하는 이유는 덱이라는 자료구조 때문
덱 편리성 메서드를 많이 제공(큐, 스텍, 회전, 최대개수, ...)

```python
history = deque(maxlen=5)   # maxlen=n: n개만 저장

history.append("google")
history.append("naver")
history.append("youtube")
history.append("github")
history.append("chatgpt")

print(history)

history.append("python")

print(history)
```

```python
cards = deque([1, 2, 3, 4, 5])

while len(cards) > 1:
    print("버림 :", cards.popleft())    #1번 카드를 버림
    cards.append(cards.popleft())
    print("남은 카드 :", cards)

print("마지막 카드 :", cards[0])
```
