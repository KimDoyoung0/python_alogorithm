실습 1. 클래스 설계하기

​

목표

- 클래스를 직접 설계한다. 

- 생성자와 속성을 이해한다. 

​

요구사항

- Product 클래스를 작성하시오.

​

- 속성

  상품명 

  가격 

  재고 

- 메서드

  총 판매금액 반환 (가격 × 판매수량) 

  상품 정보 출력 

​

실행 예

p = Product("키보드", 50000, 10)

print(p.name)
print(p.price)
print(p.stock)

```python
class Product:
    def __init__(self, name, price, stock):
        self.name = name
        self.price = price
        self.stock = stock

    def total_price(self, count):
        return self.price * count

    def print_info(self):
        print(self.name)
        print(self.price)
        print(self.stock)
```

```python
p = Product("키보드", 50000, 10)

p.print_info()
```

​

목표

- 여러 개의 상품 객체를 생성한다.

​

요구사항

- 다음 상품을 객체로 생성하시오.

키보드   50000   10
마우스   30000   20

```python
products = [
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5),
    Product("노트북", 1200000, 3),
]

#print(len(products))

for p in products:
    #p.print_info()
    if p.stock >= 10:
        #p.print_info()
        print(p.name)
```

```python
# 세일즈 목록

sales = [
    Product("키보드", 50000, 10),
    Product("키보드", 50000, 10),
    Product("마우스", 30000, 20),
    Product("모니터", 250000, 5),
    Product("노트북", 1200000, 3),
    Product("노트북", 1200000, 3),
    Product("노트북", 1200000, 3),
]

# {"키보드":2, "마우스":1, "모니터":1, "노트북":3}
counts = {}
for s in sales:
    if s.name not in counts:    # counts에 상품명 키가 없다면
        counts[s.name] = 1
    else:
        counts[s.name]+=1

print(counts)

check_name_list = []  # 같은 상품이 2번 계산되면 안 되니까 계산했으면 입력
for s in sales:
    if s.name not in check_name_list:
        count = counts[s.name]
        print(f"{s.name}: {s.total_price(counts[s.name])}")
        check_name_list.append(s.name)
```
