```python
class Student:
    def __init__(self, name):
        self.name = name
    def print_name(self):
        print(self.name)
    def greet(self):
        print(f"{self.name}님 hello")

# 상속하면
class HighStudent(Student):
    def study(self):
        print("공부한다.")
```

```python
s= HighStudent("홍길동")
s.print_name()
s.study()
s.greet()
```

```python
def average(student):
    return (student["kor"] + student["eng"]) / 2


student = {
    "name": "홍길동",
    "kor": 90,
    "eng": 85,
    "average": average
}

print(student["average"](student))

student2 = {
    "name": "김철수",
    "kor": 60,
    "eng": 35,
    "average": average
}

print(student2["average"](student2))
```

```python
class Student:

    def __init__(self, name, kor, eng, math):
        self.name = name
        self.kor = kor
        self.eng = eng
        self.math = math

    def average(self):
        return (self.kor + self.eng+ self.math) / 3


s = Student("홍길동", 90, 85, 10)
s.math = 80
s2 = Student("김철수", 40, 20, 0)
s2.math = 30

print(s.average())
print(s2.average())
```
