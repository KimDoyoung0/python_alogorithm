이렇게 누적합을 구해두면 이 누적합 배열(리스트)를 이용하여 구간 합 계산도 할 수 있다.

A배열의 1번째부터 마지막 5번째까지(인덱스는 0~4 구간)의 구간합은? 배열 P[5]의 값 

A배열의 2번째부터 마지막 5번째까지(인덱스는 1~4 구간)의 구간합은? 배열 P[5] - P[2-1]의 값

A배열의 3번째부터 마지막 4번째까지(인덱스는 2~3 구간)의 구간합은? 배열 P[4] - P[3-1]의 값

​

* 구간합 = P[end] - P[start-1] 라는 공식이 만들어 진다.

```python
nums = [10,20,30,40,50]

total = 0

#특정 구간의 합을 물어볼 때 마다 그 구간의 합을 구하는 연산을 반복하여야 한다
for i in range(0,4):
    total+=nums[i]
print(total)

total = 0
#이전에 물었던 구간과 반복구간이 있어도 다시 연산을 반복해야 한다.
for i in range(1,3):
    total += nums[i]
print(total)
```

```python
nums = [10,20,30,40,50]
prefix = [0]
for n in nums:
    prefix.append(prefix[-1]+n) 
    # 누적합 값인 원소의 마지막 원소에 다음 원소값을 누적해서 append
    # 0+10->10 append [0, 10]
    # 10+20->30 append [0, 10, 30]
    # 30+30->60 append [0, 10, 30, 60]
    # ...  [0, 10, 30, 60, 100, 150]
print(prefix)
```

```python
nums = [10, 2, 30, 21, 44, 5, 8, 50, 7]

prefix = [0]

for n in nums:
    prefix.append(prefix[-1]+n)

# 주의: 구간의 숫자는 인덱스가 아니다
# 3번째(값: 30)에서 7번째(값: 8)까지의 합
start = 3
end = 7

result = prefix[end] - prefix[start-1]

print(result)
```

```python
#학생 점수 리스트 - 첫번째 학생 점수 80, 두번째 학생 점수 90, ...
scores= [80, 90, 75, 88, 95, 70, 85]
```

- 1~3번 학생 총점은? (scores 리스트 인덱스 0~2)

- 2~5번 학생 총점은? 

- 3~7번 학생 총점은? 

- 1~7번 학생 총점은?

```python
scores = [80, 90, 75, 88, 95, 70, 85]

prefix = [0]

for s in scores:
    prefix.append(prefix[-1]+s)

print(prefix)   #[0,80,170,245,333,428,583]

#1~3번 학생 총점?
print(prefix[3] - prefix[1-1])
```

* 누적합 알고리즘은

- 구간 합을 여러 번 물어볼 것이므로, 미리 누적합을 만들어 두고 빠르게 답하기 위해 사용하는 알고리즘이다.

- 계속 같은 덧셈을 하지 않기 위해 앞에서부터의 합을 미리 저장해 두는 것이 누적합이다
