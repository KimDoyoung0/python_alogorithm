1. 팀장 번호 구하기 (산술 연산자)

1번부터 100번까지 학생이 있다.

1~10번은 1팀 

11~20번은 2팀

21~30번은 3팀 

... 

각 팀은 10명씩이며 첫 번째 학생이 팀장이다.

학생 번호를 입력받아 해당 학생의 팀장 번호를 출력하는 프로그램을 작성하시오.

실행 예)

학생 번호 : 8
팀장 번호 : 1

학생 번호 : 15
팀장 번호 : 11

```python
num = int(input("학생 번호: "))

# (num-1) // 10
# 1~10 -> 0
# leader_num = ((num-1) // 10)+1

# 11~20 -> 10
leader_num = ((num-1) // 10 * 10)+1
print(leader_num)
```

2. 홀수와 짝수 판별하기 (if~else)

정수를 입력받아 홀수인지 짝수인지 출력하는 프로그램을 작성하시오.

```python
try:
    num = int(input("숫자를 입력하세요: "))
    if num%2==0:
        print("짝수")
    else:
        print("홀수")
except ValueError as e:
    print("정수만 입력하세요.")
    #print(e)
```

3. 양수, 음수, 0 판별하기 (if~elif~else)

정수를 입력받아 양수, 음수 또는 0을 출력하는 프로그램을 작성하시오.

```python
try:
    num = int(input("숫자를 입력하세요: "))
    if num>0:
        print("양수")
    elif num<0:
        print("음수")
    else:
        print("0")
except ValueError:
    print("정수만 입력하세요.")
```

4. 20대 출입 가능 여부

두 사람의 나이를 입력받는다.
두 사람 모두 20대(20~29세)이면 출입 가능을 출력하고,
그 외에는 출입 불가를 출력하는 프로그램을 작성하시오.

```python
try:
    person1, person2 = map(int, input("사람1, 2의 나이를 입력하세요: ").split())

    if 20<=person1<=29 and 20<=person2<=29:
        print("출입 가능")
    else:
        print("출입 불가")
    
except ValueError:
    print("정수만 입력하세요.")
```

```python
try:
    ages = list(map(int, input("사람1과 2의 나이를 입력하세요: ").split()))

    if all(20<=age<=29 for age in ages):
        print("출입 가능")
    else:
        print("출입 불가")
    
except ValueError:
    print("정수만 입력하세요.")
```

7. 윤년 구하기

2014년부터 2500년까지의 윤년을 모두 출력하는 프로그램을 작성하시오.

윤년의 조건

- 400으로 나누어 떨어지는 해 

- 또는 4로 나누어 떨어지지만 100으로는 나누어 떨어지지 않는 해

```python
for year in range(2014, 2500):
    if year%400 == 0 or (year%4==0 and year%100 != 0):
        print(f"{year}은 윤년입니다.")
```

8. 소수 판별하기

정수를 입력받아 소수인지 아닌지 판별하는 프로그램을 작성하시오.

(소수는 1보다 큰 자연수 중 1과 자기 자신만으로 나누어지는 수이다.)

```python
try:
    num = int(input("정수를 입력하세요: "))
    is_prime=True

    if num>0:
        for i in range(2, num):
            if num%i == 0:
                is_prime=False
                break
    
        if is_prime:
            print(f"{num}은 소수 입니다.")
        else:
            print(f"{num}은 소수가 아닙니다.")

    else:
        print(f"{num}은 양수가 아닙니다. 양수로 입력해주세요.")

except ValueError:
    print(f"{num}은 정수가 아닙니다.")
```

9. 3~100 사이의 소수 개수 구하기

3부터 100까지의 소수의 개수를 출력하는 프로그램을 작성하시오.

```python
prime_count=0

for num in range(3,101):
    is_prime=True
    for i in range(2,num):
        if num%i == 0:
            is_prime=False
            break
    if is_prime:
        prime_count+=1

print(f"소수는 총 {prime_count}개 있습니다.")
```
