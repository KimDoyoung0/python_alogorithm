```python
import re

text = "apple banana cat"
result = re.findall(r"[abc]", text) # "apple banana cat"를 0부터 끝까지 "a", "b", "c" 중 일치하면 반환되는 리스트에 추가
print(result)
# ['a', 'b', 'a', 'a', 'a', 'c', 'a']
result2 = re.findall(r"[xyz]", text) # "apple banana cat"를 0부터 끝까지 "a", "b", "c" 중 일치하면 반환되는 리스트에 추가
print(result2)
result3 = re.findall(r"apple", text) # "apple banana cat"를 0부터 끝까지 "a", "b", "c" 중 일치하면 반환되는 리스트에 추가
print(result3)
```

```python
text = "apple banana cat"
result = re.findall(r"[at]", text)

if result:
    print("a나 t가 있다.")
else:
    print("a나 t가 없다")
```

```python
text = "apple banana cat"
result = re.search(r"[at]", text)
# search는 match(있음)나 none(없음)을 반환함

if result:
    print("a나 t가 있다.")
else:
    print("a나 t가 없다")
```

```python
import re

text = "A12B34"
print(re.findall(r"[0-9]", text))
# ['1', '2', '3', '4']
# 문자0부터 문자9까지 하나라도 일치하면 반환되는 리스트에 추가 
# 정규표현식 [0-9]는 [0123456789] 동일한 패턴
```

```python
# 영문대문자 찾기

text = "Python JAVA AI"
print(re.findall(r"[A-Z]", text))

text = "Python JAVA AI"
print(re.findall(r"[a-zA-Z]", text))
```

```python
# 문자제외 [^ ]
# 대괄호 안에 ^는 괄호안에 문자를 제외한 문자

text = "aZ7bc123@#$%"
result = re.findall(r"[^0-9]", text) # 0~9문자가 아니면 반환되는 리스트에 추가
print(result)
```

```python
# 대소문자, 숫자, 특수문자 포함해야 비밀번호 생성 가능

import re

password = input("비밀번호를 입력하세요: ")
result = []

if re.search(r"[^가-힣]",password):
    result.append("한글을 사용할 수 없습니다.")
if not re.search(f"[a-z]", password):
    result.append("소문자가 없습니다.")
if not re.search(f"[A-Z]", password):
    result.append("대문자가 없습니다.")
if not re.search(f"[0-9]", password):
    result.append("숫자가 없습니다.")
if not re.search(f"[^a-zA-Z0-9가-힣]", password):
    result.append("특수문자가 없습니다.")
if len(result) == 0:
    print("사용 가능한 비밀번호 입니다.")

for r in result:
    print(r)
```
