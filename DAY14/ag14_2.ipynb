```python
import re

text = "apple banana cat"
result = re.findall(r"[abc]", text) # "apple banana cat"를 0부터 끝까지 "a", "b", "c" 중 일치하면 반환되는 리스트에 추가
print(result)
# ['a', 'b', 'a', 'a', 'a', 'c', 'a']
result2 = re.findall(r"[xyz]", text) # "apple banana cat"를 0부터 끝까지 "a", "b", "c" 중 일치하면 반환되는 리스트에 추가
print(result2)
result3 = re.findall(r"apple", text) # "apple banana cat"를 0부터 끝까지 "a", "b", "c" 중 일치하면 반환되는 리스트에 추가
print(result3)
```

```python
# 대소문자, 숫자, 특수문자 포함해야 비밀번호 생성 가능

import re

password = input("비밀번호를 입력하세요: ")
result = []

if re.search(r"[^가-힣]",password):
    result.append("한글을 사용할 수 없습니다.")
if not re.search(f"[a-z]", password):
    result.append("소문자가 없습니다.")
if not re.search(f"[A-Z]", password):
    result.append("대문자가 없습니다.")
if not re.search(f"[0-9]", password):
    result.append("숫자가 없습니다.")
if not re.search(f"[^a-zA-Z0-9가-힣]", password):
    result.append("특수문자가 없습니다.")
if len(result) == 0:
    print("사용 가능한 비밀번호 입니다.")

for r in result:
    print(r)
```

6) 시작(^)과 끝($)

- 문자열 안에서 패턴의 문자를 찾기보다는 문자열 전체가 특정 문자열로 시작 or 끝나는 지를 확인할때 사용

```python
# findall함수에서는 ^패턴을 사용해도 의미있는 반환값을 얻을 수 없다(^패턴은 전체 문자열을 하나로 다루기 때문)
text = "Python Python python"
print(re.findall(r"^Py", text)) # ["Py"]
print("===========================")

text = "Python Python python"
print(re.findall(r"(Py)", text))
print("===========================")

text = "Python Python python"
print(re.findall(r"Py", text))
print("===========================")

text = "python Python python"
print(re.findall(r"^Py", text)) # [] 

print("===========================")
texts = ["Python Python python", "python Python python"]
for t in texts:
    print(re.findall(r"^Py", text))
```

```python
text = "Python Python python"
print(re.search(r"^Py", text)) # <re.Match object; span=(0, 2), match='Py'> span은 Py문자의 위치 0에서 2-1에 위치
print(re.search(r"^py", text)) # None
```

```python
text = "Python Python python"
print(re.match(r"Py", text)) # <re.Match object; span=(0, 2), match='Py'> span은 Py문자의 위치 0에서 2-1에 위치
print(re.match(r"py", text)) # None
print(re.match(r"Py", text).group())
#match는 시작점에서 찾고, search는 모든 문장에서 py로 시작하는걸 찾는 거임

# match(): 문자열의 앞 부분이 매치되는가를 체크, 추출
# search(): 선두에 한해서 매치하는지를 체크, 추출
```

```python
text = "Python"

print(re.search(r"hon$", text)) # <re.Match object; span=(3, 6), match='hon'>
print(re.search(r"abc$", text)) # None
print(re.search(r"on$", text))
# 끝나는걸 찾고 싶으면 search에서 $를 넣어서 찾기 ex)re.search(r"hon$", text)
```

```python
text = "Pyyyyython"

# 시작이 Py 찾기
print(re.search(r"^Py", text)) # <re.Match object; span=(0, 2), match='Py'>
# 시작이 Pyyyyy.... 찾기
print(re.search(r"^Py*", text)) # <re.Match object; span=(0, 6), match='Pyyyyy'>
```

```python
text = "Pthon"

# 시작이 Py 찾기
print(re.search(r"Py", text)) # <re.Match object; span=(0, 2), match='Py'>
# 시작이 Pyyyyy.... 찾기
print(re.search(r"Py*", text)) # <re.Match object; span=(0, 6), match='Pyyyyy'>
#P뒤에 y가 있어도 되고, 없어도 된다 (*)
print(re.search(r"Py+", text))
#P뒤에 y가 하나 이상은 있어야 한다. (+)
```

```python
texts = ["3", "123", "", "abc", "12a"]

for t in texts:
    print(t, "->", bool(re.fullmatch(r"\d+", t))) # \d+ : 숫자가 1개 이상 있어야 한다.
    # fullmatch: 전체가! 숫자여야 한다.
    print(t, "->", bool(re.fullmatch(r"\d*", t))) # \d* : 숫자가 있어도 되고, 없어도 된다
    print("===============")

"""
 -> False
 -> True
 결과 이유: 숫자가 있어도 되고, 없어도 되는데 숫자가 없으니께
 """
```

8) 수량자 ?

- 앞 문자가 0개 또는 1개 존재한다.
- 정규표현식에서 ?는 바로 앞 문자가 0번 또는 1번 등장한다는 의미

- "ab?c" : b가 없거나 1개 있거나 -> "abc"(O) 와 "ac"(O)는 매칭,  "abbc"(X) b가 2개라서 매칭 안됨

```python
words = [
    "ae",
    "abc",
    "abbcd",
    "abbbcde",
    "aabcdfs",      #match는 첫번째 문자부터 찾으므로 a->b나 c여야하는데 aa라서 none 나온거임
    "acde"
]

for word in words:
    print(re.match(r"ab?c", word))
# None
# <re.Match object; span=(0, 3), match='abc'>
# None
# None
# None
# <re.Match object; span=(0, 2), match='ac'>
```
