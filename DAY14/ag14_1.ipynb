```python
# 슬라이딩 윈도우에 사용될 리스트
prices = []

for r in sales:
    if isinstance(r["판매액"], int):
        prices.append(r["판매액"])

window = 3
# [0:3] [1:4] [2:5]
for i in range(len(prices)-(window-1)):  #range(len(price)-window+1)
    total = sum(prices[i:i+window])
    avg = round(total/window, 3)
    print(f"{i+1}~{i+window}위 평균 판매액: {avg}원")
```

```python
#결측치 처리, 결측값("", "결측")을 평균이나 중앙값으로 변경
#try 예외처리 또는 if 조건분기

avg = sum(prices)/len(prices)

prices.sort()
print(prices)
median = 0  # 중앙값 초기화

if len(prices) % 2 == 1:
    median = prices[len(prices)//2]
else:
    median = (prices[len(prices)//2] + prices[len(prices)//2-1])/2
    # median = (prices[len(prices)//2] + prices[len(prices)//2+1])/2 아님... 인덱스!라서 -1 인거임

print(median)
print(f"원래 sales: {sales}")

for r in sales:
    if not isinstance(r["판매액"], int):
        r["판매액"] = median
print("결측치->중앙값 치환: ", sales)
```

```python
result = {}

for r in sales:
    p_name = r["상품"]
    price = r["판매액"]
    total_cnt = r["수량"]

    if p_name not in result:
        result[p_name]={"합계":0, "건수":0, "최대":price, "최소":price, "총수량":0}
    
    #result[p_name]["합계"] += price
    info = result[p_name]
    info["합계"] += price
    info["건수"] += 1
    info["총수량"] += total_cnt

    #최대/최소
    if info["최대"] < price:
        info["최대"] = price
    if info["최소"] > price:
        info["최소"] = price

result
#print(result)
```
